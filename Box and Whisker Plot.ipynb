```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
```

```python
x=np.random.normal(size=10000)
x
```

```python
sns.kdeplot(x,shade=True,color='cyan')
plt.title('Standard Normal Distribution')
```

```python
fig,ax=plt.subplots(ncols=2)
fig.set_size_inches(16,6)
sns.boxplot(x,whis=x.max(),ax=ax[0],color='crimson')
ax[0].title.set_text('Whisker=Maximum Outlier Value')
sns.boxplot(x,whis=x.min(),ax=ax[1],color='gold')
ax[1].title.set_text('Whisker=Quartiles Value')

fig.suptitle('Length of Whisker',color='darkviolet',fontsize=14)
```

```python
iqr=xs.quantile(.75)-xs.quantile(.25)
print('Inter-Quartile Range\t',iqr,'or',0.667829-(-0.690435))
```

```python
l_whisker=xs.quantile(.25)-(1.5*iqr)
r_whisker=xs.quantile(.75)+(1.5*iqr)
print('Left-Whisker\t',l_whisker,'\nRight-Whisker\t',r_whisker)
```
